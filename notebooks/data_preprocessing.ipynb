```python
import os
import pandas as pd
import numpy as np
import glob
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(filepath_or_buffer="../data/raw/UNSW_NB15_training-set.csv")
```

```python
df2 = pd.read_csv(filepath_or_buffer="../data/raw/Monday-WorkingHours.csv")
```

```python
files = glob.glob("../data/raw/CICIDS2017_test/*.csv")
test_df = pd.concat([pd.read_csv(file) for file in files], ignore_index=True)
```

```python
files = glob.glob("../data/raw/CICIDS2017_test/*.csv")
cic_df = pd.concat([pd.read_csv(file) for file in files], ignore_index=True)
```

```python
train_df, val_df = train_test_split(
    df,
    test_size=0.3,
    random_state=42,
    stratify=df["attack_cat"]
)
```

```python
mon_train, mon_val = train_test_split(
    df2,
    test_size=0.3,
    random_state=42,
    shuffle=True
)
```

```python
cic_train_df, cic_temp_df = train_test_split(
    cic_df,
    test_size=0.4,
    random_state=42,
    stratify=cic_df["Label"]
)

cic_val_df, cic_test_df = train_test_split(
    cic_temp_df,
    test_size=0.5,
    random_state=42,
    stratify=cic_temp_df["Label"]
)
```

```python
train_df.to_csv("../data/processed/UNSW_NB15_train.csv", index=False)
val_df.to_csv("../data/processed/UNSW_NB15_val.csv", index=False)
```

```python
mon_train.to_csv("../data/processed/CICIDS2017_mon_train.csv", index=False)
mon_val.to_csv("../data/processed/CICIDS2017_mon_val.csv", index=False)
```

```python
test_df.to_csv("../data/processed/CICIDS2017_attack_test.csv", index=False)
```

```python
cic_train_df.to_csv("../data/processed/svCICIDS2017_train.csv", index=False)
cic_val_df.to_csv("../data/processed/svCICIDS2017_val.csv", index=False)
cic_test_df.to_csv("../data/processed/svCICIDS2017_test.csv", index=False)
```
