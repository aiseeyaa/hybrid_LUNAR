```python
import pandas as pd
import numpy as np
```

```python
df_morning = pd.read_csv(filepath_or_buffer="../data/raw/Friday-WorkingHours-Morning.pcap_ISCX.csv")
#df_portscan = pd.read_csv(filepath_or_buffer="../data/sample/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv")
#df_ddos = pd.read_csv(filepath_or_buffer="../data/sample/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")
```

```python
print(df_morning.columns.tolist())
```

```python
print(df_morning.shape)
print(df_morning[" Label"].value_counts())
print(df_morning.isna().sum().sort_values(ascending=False).head(10))
```

```python
df_morning[' Label'].value_counts()
```

```python
df_morning = df_morning.drop(['Flow ID', 'Source IP', 'Source Port', 'Destination IP',
              'Destination Port', 'Timestamp'], axis=1, errors='ignore')
```

```python
df_morning[' Label'] = df_morning[' Label'].map({'BENIGN': 0, 'Bot': 1}).astype(float)
```

```python
for col in df_morning.columns:
    if col != ' Label':
        df_morning[col] = pd.to_numeric(df_morning[col], errors='coerce')

df_morning = df_morning.replace([np.inf, -np.inf], np.nan)
df_morning = df_morning.dropna()

df_morning = df_morning[~df_morning.duplicated()].reset_index(drop=True)

feature_cols = [c for c in df_morning.columns if c != ' Label']
df_morning = pd.concat([df_morning[feature_cols], df_morning[' Label']], axis=1)
```

```python
df_morning.to_csv("../external/LUNAR/data/CICIDS/morning_bot.csv", index=False)
```
